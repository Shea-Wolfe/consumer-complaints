```python
import pandas as pd
import numpy as np
import random
import math
import statistics as st
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
complaints = pd.read_csv('complaints_dec_2014.csv')
```

```python
complaints.Product.value_counts().plot(kind='bar', title='Complaints by Product or Service')
```

# Results

* In the table above you can see the most frequent products that had complaints were Mortgage, Debt Collection, and Credit Reporting. 
* Surprisingly, payday loans are very low on the list despite being an industry with a bad reputation, suggesting perhaps a reporting bias.

```python
complaints.Company.value_counts()[:10].plot(kind='bar', title='Complaints by Company')
```

# Results

* In the graph above you can see that the complaints are spread out over a large number of financial institutions.

```python
complaints['Company response'].value_counts().plot(kind='bar', title='Complaints by Company Responce')
```

# Results

* In the graph above you can see that the most common result of a complaint was a company responce with explanation but only rarely was any sort of relief offered.

```python
dates = complaints['Date received']
```

```python
complaints.index = pd.to_datetime(dates, format='%m/%d/%Y')
```

```python
complaints['Weekdays'] = complaints.index.to_series().map(lambda d: d.weekday())
```

```python
plt.bar(range(7),[complaints[complaints.Weekdays == x].count().Product/len(complaints[complaints.Weekdays == x].index.unique()) for x in range(7)])
plt.xticks([x+.5 for x in range(7)], ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],rotation=90)
plt.title('Average Daily Complaints')
plt.ylabel('Number of complaints')
plt.show()
```

# Results

* In the graph above you can see that complaints peak on Tuesdays, but are overall much more frequent on the weekdays with drastically lower rates on Saturday and Sunday.
